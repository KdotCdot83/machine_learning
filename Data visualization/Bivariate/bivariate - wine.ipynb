```python
import pandas as pd

reviews = pd.read_csv('/Users/kcclaws/machine_learning/Data visualization/winemag_data.csv')
reviews.head()
```

```python
%matplotlib inline
reviews[reviews['price'] < 100].sample(100).plot.scatter(x='price',
                                                         y = 'points')
```

```python
# If you don't get a smaple, the chart isn't very useful
reviews[reviews['price'] < 100].plot.scatter(x='price',y='points')
```

```python
# Hexplots can get around this issue
reviews[reviews['price'] < 100].plot.hexbin(x='price',
                                            y='points',
                                           gridsize=15)
```
